```python
from pyspark.sql import SparkSession
from pyspark.sql import Window
import pyspark.sql.functions as F


input_file_path = '/content/Input_data.csv'


def process_with_pyspark():
    spark = SparkSession.builder.appName("WebTreeBuilder").getOrCreate()

    df_spark = spark.read.csv(input_file_path, header=True)

    window_spec = Window.orderBy("PARENT_CATEGORY_ID").rowsBetween(Window.unboundedPreceding, Window.currentRow)
    df_spark = df_spark.withColumn("Web_TREE", 
                                   F.concat_ws('_', 
                                               F.collect_list("PARENT_CATEGORY_ID").over(window_spec))
                                  )

    df_spark.write.json("/content/out/json", mode="overwrite")

    df_spark.write.parquet("/content/out/parquet", mode="overwrite")
    
    print("PySpark Output:")
    df_spark.show(truncate=False)

    # Stop the Spark session
    spark.stop()

if __name__ == "__main__":

    process_with_pyspark()
```
